```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lines = []
with open('onboard/problem1-1.txt',  'r') as f:
    lines = [line.split('\t') for line in f.readlines()]
    print(lines)
```

```python
header = lines[0]
start = float(lines[1][0])
body = lines[1:]
body = np.array([[float(i) for i in line[:-1]] for line in body])
body[:, 0] -= start
body
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Power over time')
ax.set_ylabel('Power(W)')
ax.set_xlabel('Time')
ax.plot(body[:, 0], body[:,1], label="Total Power")
ax.legend()
plt.show()
fig.savefig("Power-Q1-1.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Temperature over time')
ax.set_ylabel('Temp.(C)')
ax.set_xlabel('Time')
for i in range(10, 14):
    ax.plot(body[:, 0], body[:,i], label=f"core {i-6}")
ax.legend()
plt.show()
fig.savefig("temp-Q1-1.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('CPU usage over time')
ax.set_ylabel('Usage(%)')
ax.set_xlabel('Time')
for i in range(6, 10):
    ax.plot(body[:, 0], body[:,i]*100, label=f"core {i-2}")
ax.legend()
plt.show()
fig.savefig("usage-Q1-1.png")
```

```python
lines = []
with open('onboard/problem1-3_blacksholes.txt',  'r') as f:
    lines = [line.split('\t') for line in f.readlines()]
    print(lines)
header = lines[0]
start = float(lines[1][0])
body = lines[1:]
body = np.array([[float(i) for i in line[:-1]] for line in body])
body[:, 0] -= start
body= body[:2400]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Power over time(blackscholes)')
ax.set_ylabel('Power(W)')
ax.set_xlabel('Time')
print(f'energy:{np.sum(body[:,1])*0.2}')
print(np.mean(body[:,1]))
ax.plot(body[:, 0], body[:,1], label="Total Power")
ax.legend()
plt.show()
fig.savefig("Power-Q1-3_blackscholes.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Core Frequency over time(blackscholes)')
ax.set_ylabel('Frequency(khz)')
ax.set_xlabel('Time')
print(np.max(body[:, 0]))
ax.plot(body[:, 0], body[:,-1], label=f"frequency")
ax.legend()
plt.show()
fig.savefig("usage-Q1-3_blackscholes.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('max big core Temperature over time(blackscholes)')
ax.set_ylabel('Temp(C)')
ax.set_xlabel('Time')
max_big_temp = np.amax(body[:,10:14],axis=1)
print(f'mean: {np.mean(max_big_temp)}')
print(f'max: {np.max(max_big_temp)}')
ax.plot(body[:, 0], max_big_temp, label=f"Max Core Temp")
ax.legend()
plt.show()
fig.savefig("temp-Q1-3_blackscholes.png")
```

```python
lines = []
with open('onboard/problem1-3_bodytrack.txt',  'r') as f:
    lines = [line.split('\t') for line in f.readlines()]
    
header = lines[0]
start = float(lines[1][0])
body = lines[1:-1]


body = [[float(i) for i in line[:-1]] for line in body]

body=np.array(body)

body[:, 0] -= start
body= body[:750]
print(np.max(body[:, 0]))
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Power over time(bodytrack)')
ax.set_ylabel('Power(W)')
ax.set_xlabel('Time')
ax.plot(body[:, 0], body[:,1], label="Total Power")
ax.legend()
plt.show()
print(np.mean(body[:,1]))
print(f'energy:{np.sum(body[:,1])*0.2}')
fig.savefig("Power-Q1-3_bodytrack.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Core Frequency over time(bodytrack)')
ax.set_ylabel('Frequency(khz)')
ax.set_xlabel('Time')

ax.plot(body[:, 0], body[:,-1], label=f"frequency")
ax.legend()
plt.show()
fig.savefig("usage-Q1-3_bodytrack.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('max big core Temperature over time(bodytrack)')
ax.set_ylabel('Temp(C)')
ax.set_xlabel('Time')
max_big_temp = np.amax(body[:,10:14],axis=1)
print(f'mean: {np.mean(max_big_temp)}')
print(f'max: {np.max(max_big_temp)}')
ax.plot(body[:, 0], max_big_temp, label=f"Max Core Temp")
ax.legend()
plt.show()
fig.savefig("temp-Q1-3_bodytrack.png")
```
