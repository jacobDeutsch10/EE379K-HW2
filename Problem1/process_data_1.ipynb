```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lines = []
with open('onboard/problem1-1.txt',  'r') as f:
    lines = [line.split('\t') for line in f.readlines()]
    print(lines)
```

```python
header = lines[0]
start = float(lines[1][0])
body = lines[1:]
body = np.array([[float(i) for i in line[:-1]] for line in body])
body[:, 0] -= start
body
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Power over time')
ax.set_ylabel('Power(W)')
ax.set_xlabel('Time')
ax.plot(body[:, 0], body[:,1], label="Total Power")
ax.legend()
plt.show()
fig.savefig("Power-Q1-1.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Temperature over time')
ax.set_ylabel('Power(C)')
ax.set_xlabel('Time')
for i in range(10, 14):
    ax.plot(body[:, 0], body[:,i], label=f"core {i-6}")
ax.legend()
plt.show()
fig.savefig("temp-Q1-1.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('CPU usage over time')
ax.set_ylabel('Usage(%)')
ax.set_xlabel('Time')
for i in range(6, 10):
    ax.plot(body[:, 0], body[:,i], label=f"core {i-2}")
ax.legend()
plt.show()
fig.savefig("usage-Q1-1.png")
```
